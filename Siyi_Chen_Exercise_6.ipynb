# Image Segmentation

```python
height, width, sequence = img_1_t2.shape
seg_1_t2=np.zeros((height,width))
for i in range(1,height):
  for j in range(1,width):
    if slice_1_t2[i,j] <150:
      seg_1_t2[i,j]=0
    elif slice_1_t2[i,j] <300:
      seg_1_t2[i,j]=1
    elif slice_1_t2[i,j] <410:
      seg_1_t2[i,j]=2
    elif slice_1_t2[i,j] <600:
      seg_1_t2[i,j]=3      
    elif slice_1_t2[i,j] <700:
      seg_1_t2[i,j]=4
    elif slice_1_t2[i,j] <880:
      seg_1_t2[i,j]=5
    elif slice_1_t2[i,j] <1000:
      seg_1_t2[i,j]=6            
    elif slice_1_t2[i,j] <1130:
      seg_1_t2[i,j]=7            
    elif slice_1_t2[i,j] <1300:
      seg_1_t2[i,j]=8   
    else:
      seg_1_t2[i,j]=9

height, width, sequence = img_2_t2.shape
seg_2_t2=np.zeros((height,width))
for i in range(1,height):
  for j in range(1,width):
    if slice_2_t2[i,j] <110:
      seg_2_t2[i,j]=0
    elif slice_2_t2[i,j] <200:
      seg_2_t2[i,j]=1
    elif slice_2_t2[i,j] <300:
      seg_2_t2[i,j]=2
    elif slice_2_t2[i,j] <400:
      seg_2_t2[i,j]=3      
    elif slice_2_t2[i,j] <510:
      seg_2_t2[i,j]=4
    elif slice_2_t2[i,j] <600:
      seg_2_t2[i,j]=5
    elif slice_2_t2[i,j] <710:
      seg_2_t2[i,j]=6            
    elif slice_2_t2[i,j] <810:
      seg_2_t2[i,j]=7            
    elif slice_2_t2[i,j] <920:
      seg_2_t2[i,j]=8   
    else:
      seg_2_t2[i,j]=9


fig = plt.figure(figsize=(16,16))
plt.figtext(.5,.9,'Segmentation Image of T2 Protocal', fontsize=20, ha='center')

plt.subplot(2,2,1)
plt.imshow(slice_1_t2,cmap='bone')
plt.colorbar()
plt.title('Original Image of Dataset 1', fontsize=16)

plt.subplot(2,2,2)
plt.imshow(seg_1_t2,cmap='jet')
plt.colorbar()
plt.title('Segmentation of Dataset 1', fontsize=16)

plt.subplot(2,2,3)
plt.imshow(slice_2_t2,cmap='bone')
plt.colorbar()
plt.title('Original Image of Dataset 2', fontsize=16)

plt.subplot(2,2,4)
plt.imshow(seg_2_t2,cmap='jet')
plt.colorbar()
plt.title('Segmentation of Dataset 2', fontsize=16)
```

# Segmentation Level
**Dataset 1**
*   0 background
*   1 2 white matter
*   3 4 gray matter
*   5 6 7 8 9 lesion

**Dataset 2**
*   0 1 background
*   2 3 white matter
*   4 5 6 gray matter
*   7 8 9 lesion

```python
seg_precise_1_t2=np.zeros((height,width))
for i in range(1,height):
  for j in range(1,width):
    if seg_1_t2[i,j] == 0:
      seg_precise_1_t2[i,j]=0
    elif seg_1_t2[i,j] == 1 or seg_1_t2[i,j] == 2:
      seg_precise_1_t2[i,j]=1
    elif seg_1_t2[i,j] == 3 or seg_1_t2[i,j] == 4:
      seg_precise_1_t2[i,j]=2
    else:
      seg_precise_1_t2[i,j]=3   

seg_precise_2_t2=np.zeros((height,width))
for i in range(1,height):
  for j in range(1,width):
    if seg_2_t2[i,j] == 0 or seg_2_t2[i,j] == 1:
      seg_precise_2_t2[i,j]=0
    elif seg_2_t2[i,j] == 2 or seg_2_t2[i,j] == 3:
      seg_precise_2_t2[i,j]=1
    elif seg_2_t2[i,j] == 4 or seg_2_t2[i,j] == 5 or seg_2_t2[i,j] == 6:
      seg_precise_2_t2[i,j]=2
    else:
      seg_precise_2_t2[i,j]=3  

fig = plt.figure(figsize=(16,16))
plt.figtext(.5,.9,'Segmentation Image of T2 Protocal', fontsize=20, ha='center')

plt.subplot(2,2,1)
plt.imshow(slice_1_t2,cmap='bone')
plt.colorbar()
plt.title('Original Image of Dataset 1', fontsize=16)

plt.subplot(2,2,2)
plt.imshow(seg_precise_1_t2,cmap='jet')
plt.colorbar()
plt.title('Segmentation of Dataset 1', fontsize=16)

plt.subplot(2,2,3)
plt.imshow(slice_2_t2,cmap='bone')
plt.colorbar()
plt.title('Original Image of Dataset 2', fontsize=16)

plt.subplot(2,2,4)
plt.imshow(seg_precise_2_t2,cmap='jet')
plt.colorbar()
plt.title('Segmentation of Dataset 2', fontsize=16)
```
